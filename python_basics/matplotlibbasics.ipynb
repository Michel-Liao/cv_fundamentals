```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3, 4]
y = [0,2,4,6,8]

# Resize graph
plt.figure(figsize=(5,3), dpi=100)
# dpi is dots per inch, but think of it as pixels per inch
# higher dpi is better
# dpi around 300 is good

# Keyword Argument Notation
# plt.plot(x,y, label='2x', color='red', linewidth=3, marker='.', markersize=10, markeredgecolor='blue', linestyle='--')

# Shorthand notation
# fmt = '[color][marker][line]'

plt.plot(x,y, 'r.-', label='2x')

# Line 2

# Select interval to plot points at
x2 = np.arange(0,4, 0.5)

# Plot part of graph as line
plt.plot(x2[:5], x2[:5]**2, 'b', label='x^2')

# Plot remainder of graph
plt.plot(x2[4:], x2[4:]**2, 'b--')

# Add a title
plt.title('First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# x and y labels
plt.xlabel('x axis!')
plt.ylabel('y axis!')

plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,7,8,10])
# graph auto resizes based on your ticks

plt.legend()

plt.savefig('mygraph.png', dpi=300)
# saves in the current directory that this file is in

# Show plot
plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

patterns = ['/', 'o', '*']

for bar in bars:
    bar.set_hatch(patterns.pop(0))

# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')

plt.figure(figsize=(6,4))

plt.show()
```

# Real World Examples

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.title('Gas Prices over Time (in USD)')

plt.plot(gas.Year, gas.USA, 'b.-', label='United States')
plt.plot(gas.Year, gas['Canada'], 'r.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')
plt.plot(gas.Year, gas.Australia, 'y.-', label='Australia')


plt.xlabel('Year')
plt.ylabel('USD')


plt.xticks(gas.Year[::3])

plt.legend()


plt.show()
```
